```python
mar_status=pd.read_csv('/Users/bethanyzhang/Desktop/marital_status_01.csv')
mar_status.head(5)
```

```python
mar_status['age'].fillna(37.2, inplace=True)
mar_status.head(5)
```

```python
mar_status['anniversary year'].fillna(0, inplace=True)
mar_status.head(5)
```

```python
mar_status['high_school_year'].fillna(1994, inplace=True)
mar_status.head(5)
```

```python
mar_status['age_group'] = mar_status['age_group_inputation'].astype('category')
mar_status['higher_pos'] = mar_status['higher_position'].astype('category')
mar_status['mar'] = mar_status['marriage'].astype('category')
mar_status['ann'] = mar_status['ann_disclose'].astype('category')
mar_status['gender_imp'] = mar_status['gender_imputation'].astype('category')
mar_status['marr_dis'] = mar_status['marr_disclose'].astype('category')
mar_status.head(5)
```

```python
from sklearn.linear_model import LogisticRegression
Y=mar_status['mar']
cols=["gender_imp","age","age_group","higher_pos","ann",'marr_dis','high_school_year']
X=mar_status[cols]
#mar_status.isnull().sum(axis=0)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
Y_test.shape
```

```python
lr= LogisticRegression()
lr.fit(X_train,Y_train)
y_pred = lr.predict(X_test)
lr.score(X_test, Y_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
confusion_matrix(Y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
y_pred1 = dt.predict(X_test)
confusion_matrix(Y_test, y_pred1)
```
